#  Number Of Miles Per Day

## First the miles are stored in a tuple, then per day miles is calculated and graph is plotted

```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd

milesCount = [55,120,170,225,265,315,350,365]
p2=list()
p2.insert(0,milesCount[0])

i=1
while i+1 <= len(milesCount):
    m=milesCount[i]
    n=milesCount[i-1]
    p2.insert(i,m-n)
    i+=1

s=pd.Series(p2)
s.index=(['1','2','3','4','5','6','7','8'])

s.plot.bar(figsize=(25,15))
plt.xlabel('Day')
plt.ylabel('Miles')
plt.suptitle('Number of Miles Covered per Day')
```
