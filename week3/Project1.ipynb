# Reading the data from file and displaying

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)

data= pd.read_csv('data.csv',index_col=0)
data
```

## Selecting delayed flights 

```python
delayed_flights = data[data['Status'] == 'delayed']
delayed_flights
```

## Bar graph displaying delayed flights city wise

```python
delayed_flights.plot.bar(figsize=(25,15))
plt.xlabel('Airline')
plt.ylabel('Delayed Flights')
plt.suptitle('Number Of Delayed Flights City wise')
```

## Count of Delayed flights of each airline

```python
r=delayed_flights.sum(axis=1)
r
```

## Comparing the total delayed flights - Bar Graph

```python
r.plot.bar(figsize=(15,5))
plt.xlabel('Airline')
plt.ylabel('Delayed Flights')
plt.suptitle('Number Of Delayed Flights')
```

## Comparing the total delayed flights - Pie Chart

```python
 plot = r.plot.pie(figsize=(5, 5))
```
