1. What is the northernmost airport in the United States?

```python
airportsData = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv')
airportsData.head()
```

To find the northernmost airport, it has to have highest latitude. TO find the highest latitude, need to sort the airport data in descending order based on latitude. 

```python
airportsData.sort_values(by='lat', ascending=False)
```

```python
Based on the above sorted data,northern most airport is Wiley Post Will Rogers Memorial but the timezone is not defined,
so one cannot determine whether it is in America or not.
```

# Hence the answer for northern most airport is Wiley Post Will Rogers Mem (BRW)

2. What is the easternmost airport in the United States?

To find the easternmost airport, it has to have highest longitude.
To find the highest longitude, need to sort the airport data in descending order based on longitude.

```python
airportsData.sort_values(by='lon', ascending=False)
```

# Based on the above data, Eastport Municipal Airport (EPM) is the easternmost airport in the United States 

3. On February 12th, 2013, which New York area airport had the windiest weather?

```python
weatherData = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv')
weatherData.head()
```

Need to filter the data based on month , year and day. Then need to sort descending order based on the wind_speed

```python
windiestWeather = weatherData[(weatherData.month == 2) & (weatherData.day == 12) & (weatherData.year == 2013)]
windiestWeather.sort_values('wind_speed', ascending = False).head()
```

# Based on the above EWR aiport had the windiest weather on February 12th, 2013
