```python
crimeData = pd.read_csv('https://raw.githubusercontent.com/Trishita-Nath/repo/master/Project%202/crime.csv')
crimeData
```

## Murders over the years sorted in descending order

```python
crimeData = crimeData.sort_values('Murder', ascending=False)
crimeMurderData = crimeData[['Year','Population','Murder']]
crimeMurderData
```

## Top 5 Years with highest robbery cases

```python
crimeData = crimeData.sort_values('Robbery', ascending=False)
crimeMurderData = crimeData[['Year','Population','Robbery']]
crimeMurderData.head()
```
